```python
import matplotlib.pyplot as plt
import pandas as pd

# great recesion and the housing market crash
EVENTS = {
    "Great Recesion" : pd.to_datetime("2008"),
    "Covid" : pd.to_datetime("2020"),
    "Dot com bubble" : pd.to_datetime("2000"),
}

ALPHA = 0.5
FONTSIZE = "9"

hdi = pd.read_csv("Data/Prepped/hdi_usa.csv", sep=';')
hdi.date = pd.to_datetime(hdi.date)

gdp = pd.read_csv("Data/Prepped/united-states-gdp-gross-domestic-product.csv")
gdp.date = pd.to_datetime(gdp.date)

pop = pd.read_csv("Data/Prepped/united-states-population-2023-11-04.csv")
pop.date = pd.to_datetime(pop.date)

une = pd.read_csv("Data/Prepped/united-states-unemployment-rate.csv")
une.date = pd.to_datetime(une.date)

cri = pd.read_csv("Data/Prepped/united-states-murder-homicide-rate.csv")
cri.date = pd.to_datetime(cri.date)

inf = pd.read_csv("Data/Prepped/united-states-inflation-rate-cpi.csv")
inf.date = pd.to_datetime(inf.date)
```

```python
def plot_the_events():
    plt.axvline(x = EVENTS["Dot com bubble"], color = "#85581d", linestyle = '--', label = "Dot-com bubble", alpha = ALPHA)
    plt.axvline(x = EVENTS["Great Recesion"], color = "#2aa89c", linestyle = '--', label = "The Great Recesion", alpha = ALPHA)
    plt.axvline(x = EVENTS["Covid"], color = "#ff00ff", linestyle = '--', label = "Covid - 19 pandemic", alpha = ALPHA)
```

```python
fig = plt.figure()
ax = plt.subplot(111)
plot_the_events()
ax.plot(gdp["date"], gdp["GDP"])
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=5, fontsize = FONTSIZE)

plt.ylabel("Billions of USD")
plt.title("USA GDP 1990-2021")
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
plot_the_events()
ax.plot(hdi.date, hdi.hdi)
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=5, fontsize = FONTSIZE)

plt.title("USA HDI 1900-2021")
plt.ylabel("HDI")
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
plot_the_events()
ax.plot(une.date, une["Unemployment Rate"])
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=5, fontsize = FONTSIZE)

plt.title("USA unemployment rate 1990-2021")
plt.ylabel("Unemployment [%]")
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(inf.date, inf["Inflation Rate"])
plot_the_events()
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=5, fontsize = FONTSIZE)
plt.ylabel("Inflation rate [%]")
plt.title("Inflation in the USA 1990 - 2020")
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(cri.date, cri.Rate)
plot_the_events()
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=5, fontsize = FONTSIZE)
plt.ylabel("Per 100k Population")
plt.title("Murder/Homicide rate in the USA 1990 - 2020")
plt.show()
```
